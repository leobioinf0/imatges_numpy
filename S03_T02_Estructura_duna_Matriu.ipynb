### [Broadcasting](https://numpy.org/doc/stable/user/theory.broadcasting.html#array-broadcasting-in-numpy)

Numpy aplica las reglas del `broadcasting` cuando las los arrays no tienen la misma forma.  

1.   **Primera regla del broadcasting**: Se añade una dimensión de 1 al principio del array con menor rango hasta que éstos coincidan.
2.   **Segunda regla del broadcasting**: Si la forma de los dos arrays no coincide en alguna dimensión, el array con forma igual a 1 en esa dimensión se repetirá para que coincida con la forma del otro array.
3.   **Tercera regla del broadcasting**: Después de aplicar las reglas 1 y 2, las dimensiones de los arrays deben coincidir. Si no es así, la operación no se puede llevar a cabo.

#### Escalar a 1-d array
Broadcasting de un escalar para poder operar con un arrar.

![scalar_to_1darray](https://docs.google.com/drawings/d/e/2PACX-1vTPfbu6sKAWGW-QX4BbCAzwSdYrr814T71JZOiSTgiY5-a2CrjOmsq2a0BSohzrx82waz685oUS8gYJ/pub?w=445&h=137)

```python
a = np.array([1, 2, 3])
print("a:\n", a)
print("a shape:\n", a.shape, "\n")

b = 2
print("b:\n",b, "\n")

print("Broadcast:\n",np.broadcast_arrays(a, b), "\n")

axb = a * b 
print("a * b:\n",axb)
print("a * b shape:\n",axb.shape)
```

#### 1-d array a 2-d array

![1darray_to_2darray](https://docs.google.com/drawings/d/e/2PACX-1vQ_4UzrxK4-YLq28do18P76Dw9R6BHfXEiqyTaHc08wBtl43zyhK2v0LWfY5iU8h4zmgAOa5WaPg11Q/pub?w=473&h=208)
